```python
import pandas as pd

reg_ids = [1, 2, 3, 4]
names = ["Andrew", "Bob", "Charlie", "david"]

data = {"reg_id": reg_ids, "name": names}

df = pd.DataFrame(data)

log_ids = [1, 2, 3, 4]
nams = ["xavier", "Andrew", "Yauren", "Bob"]

fd = pd.DataFrame({"log_id": log_ids, "name": nams})

fd

```

```python
pd.merge(df, fd, on="name")
```

```python
df[df["name"].isin(fd["name"])]
```

```python
fd[fd["name"].isin(df["name"])]
```

```python
pd.merge(df, fd, on="name", how= "left")
```

```python
not_logged_in = df[~df['name'].isin(fd['name'])]
num_not_logged_in = len(not_logged_in)
names_not_logged_in = not_logged_in['name'].tolist()

print(f"Number of people who have registered but not logged in: {num_not_logged_in}")
print(f"Names of people who have registered but not logged in: {names_not_logged_in}")
```

```python
not_registered = fd[~fd['name'].isin(df['name'])]
names_not_registered = not_registered['name'].tolist()

print(f"Number of people who have attended the event but not registered: {len(not_registered)}")
print(f"Names of people who have attended the event but not registered: {names_not_registered}")
```
